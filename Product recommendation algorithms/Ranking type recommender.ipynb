```python
medias = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/medias.csv')
```

```python
medias.sort_values('media_rating', ascending=False)
```

```python
corte_votos= medias['media_rating'].quantile(0.80)
corte_votos
```

```python

# Filtrado de películas según el punto de corte
medias_filtrado = medias[medias.conteo >= corte_votos].reset_index()

# Cálculo de C
media = medias_filtrado.media_rating.mean()

# Definición de la función para calcular el score
def media_imdb(x, m=corte_votos, C=media):
    v = x['conteo']
    R = x['media_rating']
    # Calculamos el weighted score
    return round((R*v + C*m) / (v+m), 2)

# Cálculo del score para todas las películas
medias_filtrado["score"] = medias_filtrado.apply(media_imdb, axis=1)

# Ordenamiento de las películas según su score
medias_filtrado = medias_filtrado.sort_values('score', ascending=False)

# Impresión del top 25 de películas recomendadas
top25 = medias_filtrado[['imdb_id','imdb_title', 'conteo', 'media_rating', 'score', 'genre']]
top25.shape
```
